```python
text = """
12月24日，佟丽娅在绿洲更新圣诞美照，并配文：“平安夜快乐，大家吃到甜甜的苹果了吗？” 照片中的她手拿平安果，身穿温暖浅色系毛衣，一脸温婉笑容直视镜头，优雅美好，气质满分。
"""
```

## Define a function. Input is a word, output is some similar words.

```python
cut_file = open("D:\\new开课吧人工智能nlp\\jupyters_and_slides\\2019-summer\\article_9k_cut.txt", 'w')
```

```python
def cut(string): return ' '.join(jieba.cut(string))
```

```python
cut('这是一个测试')

import re

def token(string):
    return re.findall(r'[\d|\w]+', string)

token('这是一个测试\n\n\n')
```

```python
line_number = 5000
for i, line in enumerate(open("D:\\new开课吧人工智能nlp\\jupyters_and_slides\\2019-summer\\article_9k.txt", encoding='utf-8')):
    if i % 100 == 0: print(i)
    cut_file.write(cut(line) + '\n')
```
